```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

Declares the data frame variable

```python
df = pd.DataFrame(data=[[1,3.2,16.9],[1,2.9,15.6],[1,1.9,11.2],[1,3.5,14.5],[1,1.1,12.2],[1,2.3,13.7]], columns=['x0', 'x1', 'Target'])
```

Fits the data to the model


```python
lr = LinearRegression()
lr.fit(df.drop('Target', 1), df.Target)
print(f'Coefficients/Weights used: {lr.coef_}')
```

Makes a prediction on the value of the test variable using the Linear Regression model

```python
test = np.array([1, 3.6]).reshape(1, -1)
print(f'Prediction made for data {test}: {lr.predict(test)}')
```

```python
print(f'Parameters for model: {lr.get_params()}')
```
