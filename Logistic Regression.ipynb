```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Randomly generated dataset

```python
df = pd.DataFrame(data=[[70,39,1,1], [17,1,7,1], [44,30, 10, 1], [69,26,2,1], [43,9,9,1], [66,27,6,1], [74,27,1,1], [118,28,114,0], [120,59,112,0], [115,44,112,0], [128,26,104,0], [54,44,114,0], [68,60,105,0], [133,54,106,0]], columns=['x0', 'x1', 'x2', 'Target'])
```

Splits the data frame into x features and target (y) features

```python
X = df.drop('Target', 1)
y = df.Target
```

Splits the dataset to get a set of training data

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=2)
```

Fits a logistic regression model to the data

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
#Prediction of test set
y_pred = lr_model.predict(X_test)
print (y_pred)
```

```python
#Actual value and the predicted value
a = pd.DataFrame({'Actual value': y_test, 'Predicted value':y_pred})
a.head()
```

```python
from sklearn import metrics 
from sklearn.metrics import classification_report, confusion_matrix
matrix = confusion_matrix(y_test, y_pred)

sns.heatmap(matrix, annot=True, fmt="d")
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
print(classification_report(y_test, y_pred))
```
