The cost function below computes J(w,b) by summing the squared errors of each training example.


Screenshot_8.png

```python
def cost_function(x,y,w,b):
  m = len(x)
  sum_of_errors =0
  for i in range(m-1):
    f_wb = w*x[i]+b
    sq_err = (f_wb - y[i])**2
    sum_of_errors += sq_err
  cost = (1/2*m)* sum_of_errors
  return cost
```

The function below computes the gradient for the parameters w and b by finding the partial derivative of the cost function with respect to w, and w.r.t. b.

Screenshot_9.png

```python
def compute_gradient(x,y,w,b):
  m = len(x)
  w_gradient = 0
  b_gradient = 0
  for i in range (m-1):
    f_wb = w*x[i]+b
    w_gradient += ((f_wb - y[i]) * x[i])
    b_gradient += (f_wb - y[i])
  b_gradient = (1/m)*b_gradient
  w_gradient = (1/m)*w_gradient
  return w_gradient, b_gradient
```

The function below implements the gradient descent algorithm by updating the parameters **w** and **b** with each iteration. The parameters are updated based on the learning rate **alpha** multiplied by each parameter's gradient. The algorithm will run for **num_iters** iterations.

Screenshot_10.png

```python
def gradient_descent(x,y,w,b,alpha,num_iters):
  j_h =[] # cost function history
  w_h = [] #w,b history

  for i in range (num_iters):
    dw, db = compute_gradient(x,y,w,b)
    w = w - alpha * dw
    b = b - alpha * db
    j_h.append(cost_function(x,y,w,b))
    w_h.append([w,b])
    if i%1000 == 0: # Prints the iteration number every 1000 iterations
      print(f"Iteration: {i} \nCost: {j_h[i]} \nw,b: {w,b}")

  return w, b, j_h, w_h
```

Parameters **w** and **b** are initialized. The number of iterations as well as the learning rate **alpha** are also specified.

```python
w = 0
b = 0

iterations = 30000
alpha = 0.0001
w, b, j_h, w_h = gradient_descent(x,y,w,b, alpha, iterations)
print(f"Gradient Descent complete. Parameters [w,b] are {w,b}")
```

Used matplotlib to plot the Cost function over iterations (this code was taken from coursera)

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12,4))
ax1.plot(j_h[:100])
ax2.plot(1000 + np.arange(len(j_h[1000:])), j_h[1000:])
ax1.set_title("Cost vs. iteration(start)");  ax2.set_title("Cost vs. iteration (end)")
ax1.set_ylabel('Cost')            ;  ax2.set_ylabel('Cost')
ax1.set_xlabel('iteration step')  ;  ax2.set_xlabel('iteration step')
plt.show()
```
